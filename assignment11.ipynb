## Show All Employess With A Total Pay Between 50,000 and 75,000

```python
q = ''' select EmployeeName,totalpay from salaries 
      where TotalPay between 50000 and 75000 
'''
sql_query(q)
```

## Show All Employees With A BasePay less than 50,000 Or A TotalPay Greater Than 100,000

```python
q = ''' select EmployeeName,basepay,totalpay from salaries
    where basepay < 50000 
    or
    Totalpay > 100000
'''
sql_query(q).head(3)
```

```python
q = ''' select EmployeeName,totalpay,basepay from salaries
    where basepay < 50000 
    and
    Totalpay > 100000
'''
sql_query(q).head(3)
```

## 9. Show All Employees With A TotalPay Benefits Value Between 125,000 and 150,000 And A Job Title Containing Word 'Director'¶

```python
q = '''select employeename,TotalPayBenefits,jobtitle from salaries
     where totalpaybenefits between 125000 and 150000
     and 
     jobtitle like '%Director%'
    
'''
sql_query(q).head(10)
```

## Show All Employees Ordered By Their TotalPay Benefits In Descending Order

```python
q = '''select EmployeeName ,TotalPaybenefits from salaries
        order by totalpaybenefits desc limit 20
'''
sql_query(q)
```

## How many employees in the salaries table have OvertimePay equal to 0?

```python
q = ''' select count(employeename) as nnumber_of_employee from salaries 
  where overtimepay = 0
  '''
sql_query(q)
```
